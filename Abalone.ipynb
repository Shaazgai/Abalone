# Type of Feature

The features of given data frame are quantitative (numeric). Letâ€™s see mode of features to define their type.

```python
dat = abalone

def Mode(x):
    a = pd.Series(x).value_counts()
    return a.idxmax()

mode_result = Mode(dat['Sex'])

mode_result
```

```python
mode_result = Mode(dat['Length'])

mode_result
```

```python
mode_result = Mode(dat['Diameter'])

mode_result
```

```python
mode_result = Mode(dat['Height'])

mode_result
```

```python
mode_result = Mode(dat['Whole weight'])

mode_result
```

```python
mode_result = Mode(dat['Shucked weight'])

mode_result
```

```python
mode_result = Mode(dat['Viscera weight'])

mode_result
```

```python
mode_result = Mode(dat['Shell weight'])

mode_result
```

```python
mode_result = Mode(dat['Rings'])

mode_result
```
